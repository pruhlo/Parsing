```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

def make_req(url, page):
    headers = {
        'User-Agent': 'Mozilla/5.0',
        'page':f'{page}'}
    resp = requests.get(url, headers=headers)
    return resp
```

```python
url = f'https://smallpacking.agrosem.ua/'
response = make_req(url, 1)
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
base_select_tags = 'div.item'

list_tags = [i.get('content') for i in soup.select(f'{base_select_tags}')]
name_topics = [i.text for i in soup.select(f'{base_select_tags}')]

d_link = dict(zip(list_tags, name_topics))
# d_link
```

```python
df = pd.DataFrame(columns=['id', 'name', 'weight', 'price'])

def get_data(df):
    base_select_tags = 'div.item' # базовий тег для отримання даних
    for page in range(1,30):
        url = 'https://smallpacking.agrosem.ua/'.format(page) # URLs сторінок з данними
        response = make_req(url, page) # результат запиту
        soup = BeautifulSoup(response.text, 'html.parser') # об'єкт супу
        list_tags = [i.get('content') for i in soup.select(f'{base_select_tags}')] # витягаємо контент
        for _ in list_tags:
            dicty = eval(_) # отримуємо словник
            df.loc[len(df.index)] = dicty # додаємо рядок до датафрейму
    return df
```

```python
df = get_data(df)
df
```

Три параметра `drop_duplicates` є необов'язковими, а нижче описано докладніше:

`keep`: цей параметр має три різні значення: `First`, `Last` і `False`. Значення `First` зберігає перше входження і видаляє останні дублікати, значення `Last` зберігає тільки останнє входження і видаляє всі попередні дублікати, а значення `False` видаляє всі повторювані рядки.
`subset`: мітка, використовувана для ідентифікації повторюваних рядків
`inplace`: містить дві умови: `True` і `False`. Цей параметр видалить повторювані рядки, якщо для нього встановлено значення `True`.

```python
df.drop_duplicates(keep='last').sort_values(by='id')
```
